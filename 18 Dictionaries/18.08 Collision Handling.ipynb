## 2) Open Addressing: hi(a) = h.f(a) + f(i)

For f(i) ---> f(0) must be zero

## Three techniques for Open Addressing: 
    
### i) Linear Probing: f(i) --> i : Find the hash function of key and go linearly till you find the empty position

![download%20%281%29.png](img/download%20%281%29.png)


    
### ii) Quadratic Probing: f(i) --> i**2 : If you want to find the empty position as fast as possible    


![download%20%282%29.png](img/download%20%282%29.png)


        
### iii) Double Hashing: f(i) --> i* h'(i) : You're using different hash function for f(i) 



In real scenario, seperate chaining works absolutely fine. In terms of time complexity, it is as good as any open addressing technique. It is easy to implement and understand.
